4. Multiplication (dot product)

```python
def matrix_multiplication(mat1,mat2):
    # get the no of rows and colums for both matrixes
    row_mat1=len(mat1)
    cols_mat1=len(mat1[0])
    row_mat2=len(mat2)
    cols_mat2=len(mat2[0])

    # Chek if multiplication is passible
    if cols_mat1 != row_mat2:
        raise ValueError("Number of columns in mat1 must be equal to munber of rows in mat2")
    # Initialize the result mat with zero to store values
    result =  [[0 for _ in range(cols_mat2)] for _ in range(row_mat1)]

    # Perform mat multiplication using method for loop
    for i in range(row_mat1):
        for j in range(cols_mat2):
            for k in range(cols_mat1):
                result[i][j] += mat1[i][k] * mat2[k][j]
    return result

# Example usage:
mat1 = [
    [1,2,3],
    [4,5,6]
]
mat2 = [
    [7,8],
    [9,10],
    [11,12]
]

result = matrix_multiplication(mat1,mat2)

# printing the result
print("Resultant Matrix:")
for row in result:
    print(row)
```

8.  Determinations


```python
A = np.array([
            [1,2],
            [3,4]])
det_A = np.linalg.det(A)
print("Determinat of A: ", det_A)
```

9. Inverse of a matrix

```python
try:
    inv_A = np.linalg.inv(A)
    print("Invetse of A:\n", inv_A)
except np.linalg.LinAlgError:
    # Singular matrix is a matrix with det() = 0
    print("Matrix A is singulat and not have an inverse.")
```

10. Eigenvalues and Eigenvectors

```python
eigenvalues , eigenvectors = np.linalg.eig(A)
print("Eigenvalues of A:\n",eigenvalues)
print("Eigenvectors of A:\n",eigenvectors)
```

11. Solving a System of linear equation (Ax=b)

```python
b = np.array([5,7])

#Solve the System
solution = np.linalg.solve(A,b)
print("MAtrix A:\n ", A)
print("Vector b:\n ", b)
print(f"Solutiion of the System Ax = b where b = {b}:\n",solution)
```
