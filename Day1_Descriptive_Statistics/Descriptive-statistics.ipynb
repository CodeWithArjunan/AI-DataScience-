# DESCRIPTIVE STATISTCS
    Descriptive statistics is a core concept in data science. It is used to summarize, organize, and describe the main features of a dataset in a simple and meaningful way.
Before applying machine learning or advanced analysis, data scientists always start with descriptive statistics to understand the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# LOAD DATASET

```python
df = pd.read_csv("2022_world_cup_squads.csv", encoding="latin1")
print(df.head())
print(df.columns)
```

# Types of Descriptive Statistics
    Descriptive statistics is divided into two major categories:

     1️⃣ Measures of Central Tendency:
    a) Mean (Average) -- Sum of all values divided by number of values.
    b) Median -- The middle value when data is sorted.
    c) Mode -- The value that occurs most frequently.

# MEAN(AVETRAGE)

```python
print("Mean Age: ", df["Age"].mean())
print("Mean Goals: ", df["Goals"].mean())
print("Mean Caps: ", df["Caps"].mean())
```

# MEDIIUM

```python
print("Medium Age: ", df["Age"].median())
print("Medium Goals: ", df["Goals"].median())
print("Medium Caps: ", df["Caps"].median())
```

# MODE

```python
print("Mode Age: ", df["Age"].mode())
print("Mode Goals: ", df["Goals"].mode())
print("Mode Caps: ", df["Caps"].mode())
```

    2️⃣ Measures of Dispersion (Spread):
    -- These describe how spread out the data is.

    a) Range -- Difference between maximum and minimum.
    b) Variance -- Measures how far each value is from the mean.
    c) Standard Deviation -- Square root of variance.

# VARIANCE

```python
print("Age Variance: ", df["Age"].var())
print("Goals Variance: ", df["Goals"].var())
print("Caps Variance: ", df["Caps"].var())
```

# STANDARD DEVIATION

```python
print("Age std dev: ",df["Age"].std())
print("Goals std dev: ",df["Goals"].std())
print("Caps std dev: ",df["Caps"].std())
```

# RANGE

```python
Age_range= df["Age"].max()- df["Age"].min()
Goals_range= df["Goals"].max()- df["Goals"].min()
caps_range= df["Caps"].max()- df["Caps"].min()

print("Age Range: " , Age_range)
print("Goals Range: " , Goals_range)
print("Caps Range: " , caps_range )
```
