# 1) Problem Definition

- Predict whether a student will pass or fail:

  Input  ‚Üí hours_studied, attendance  
  Output ‚Üí pass (1) or fail (0)

# 2)Data Collection

```python
import pandas as pd

data = {
    "hours_studied": [1, 2, 3, 4, 5, 6, 7, 8],
    "attendance": [40, 50, 55, 60, 65, 70, 80, 90],
    "result": [0, 0, 0, 0, 1, 1, 1, 1]  # 0 = Fail, 1 = Pass
}

df = pd.DataFrame(data)
df
```

# 3) Data Preprocessing

```python
x = df[["hours_studied","attendance"]]  # Features
y = df["result"]
```

# 4) Exploratory Data Analysis(EDA)

```python
# Understand the Data Structure
print("Understand the Data Structure: ")
print(df.head())
print(df.tail())
print(df.shape )
print(df.info())

# Identify Missing Values
print("Identify Missing Values: ")
print(df.isnull().sum())

# Understand Statistical Summary
print("Understand Statistical Summary: ")
print(df.describe())

# Analyze Categorical Data
print("Analyze Categorical Data: ")
print(df['result'].value_counts())
print(df['result'].value_counts(normalize=True))

# Detect Outliers
print("Detect Outliers: ")
import matplotlib.pyplot as plt

plt.boxplot(df['attendance'])
plt.show()

# Relationship Between Features
print("Relationship Between Features: ")
print(df.corr())
```

# 6) Model Selection
‚ÄúThis is a classification problem because the target variable is categorical (Pass/Fail). When the output is a class or label rather than a continuous numerical value, we use classification algorithms.‚Äù

```python
#| Problem Type   | Model                  | Case                          |
#| -------------- | ---------------------- | ----------------------------- |
#| Regression     | Linear Regression      | Output is a number            |
#| Classification | Logistic Regression    | Output is a category / class  |
#| Complex        | Random Forest, XGBoost |                               |
```

# 7) Model Training
Split data into train & test

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train, x_test, y_test, y_train = train_test_split(
    x, y, test_size=0.25,random_state=42
    )

model = LogisticRegression()
model.fit(x_train, y_test)
print(model)
```

# 8) Model Evaluation
    Check how good the model is:

```python
from sklearn.metrics import accuracy_score, confusion_matrix

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

# 9) Model Tuning
    - Improve performance by:
 - Changing hyperparameters
 - Trying another algorithm
 - Adding better features

-- Here, we are not changing data:
     
- We are not changing algorithm 
- We are changing how the algorithm behaves

-- Other Important Hyperparameters in Logistic Regression
 1. penalty -- Type of regularization
 2. solver -- Algorithm used to optimize
 3. max_iter -- Number of training iterations
 4. class_weight -- Handles imbalanced data
5. fit_intercept -- Adds bias term or not

```python
# model = LogisticRegression(c=0.5)
model = LogisticRegression(
    C=0.5,
    penalty='l2',
    solver='liblinear',
    max_iter=1000
)
model.fit(x_train, y_test)
```

# 10) Deployment & Monitoring
Deploy model:
 - Web app (Flask / FastAPI)
 - Mobile app
 - Cloud API

```python
new_std = [[6,75]]  # hours studied , attendance
prediction = model.predict(new_std)

print("Pass" if prediction[0] == 1 else "Fail")
```
