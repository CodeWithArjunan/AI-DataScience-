# Data Leakage
         Data leakage happens when information from outside the training process is accidentally used to train the model, giving unrealistically high performance.

    ‚ÄúData leakage occurs when information from the test set or future data is unintentionally used during model training, leading to overly optimistic performance that fails in real-world scenarios.‚Äù

    Imagine an exam:
   - You train using question paper + answer key
   - In mock tests, you score 100%
   - In real exam ‚Üí ‚ùå fail
   - That‚Äôs data leakage.

# 1Ô∏è‚É£ Target Leakage:
    
When a feature directly or indirectly contains the target information.

```python
import pandas as pd

data = {
    "hours_std": [2,4,6,8],
    "attendance": [50,60,75,90],
    "final_result": [0,0,1,1],
    "passed_exam": [0,0,1,1]  # LEAKAGE
}

df = pd.DataFrame(data)

x = df.drop("passed_exam", axis=1)
y = df["passed_exam"]
```

# 2Ô∏è‚É£ Train‚ÄìTest Leakage:
When test data influences training.
* Scaling before splitting data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# # üß™ Wrong Way (Leakage)

# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(x)  # ‚ùå Uses full data

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Correct Way (No Leakage)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)  # fit only on train
X_test = scaler.transform(X_test)        # transform test
```

# How to Detect Data Leakage?
    Ask yourself:
    - Is this feature available at prediction time?
    - Am I using future data?
    - Did I preprocess before splitting?
    - Is accuracy ‚Äútoo good to be true‚Äù?
