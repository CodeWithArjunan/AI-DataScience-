# 2Ô∏è‚É£ Types of Probability (Used in ML)
üîπ 1. Marginal Probability

Probability of a single event

	P(A)

Example:
Probability that a student passes = 0.7

üîπ 2. Joint Probability

Probability of two events together

	P(A‚à©B)

Example: Student passes AND studies daily

üîπ 3. Conditional Probability ‚≠ê (Very Important)

Probability of A given B happened

	P(A‚à£B)=P(A‚à©B)/P(B)

Example: Probability of passing given student studied
