# PROBABILITY DISTRIBUTION

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("2022_world_cup_squads.csv" , encoding="latin1")
print(df.head())
```

# PROBABILITY

```python
avg_age=df["Age"].mean()
print(avg_age)
```

# CONDITIONAL PROBABILITY

```python
Positions=df[df["Position"]== 'Forward']
Ages=Positions['Age'].mean()
print(Ages)
```

# NORMAL DISTRIBUTION

```python
plt.figure()
plt.hist(df["Age"] , bins=30 , density=True)
plt.title("Normal Distribution")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

# MEAN & STD DEV

```python
mean_caps=df['Caps'].mean()
std_dev=df["Caps"].std()
print(mean_caps)
print(std_dev)
```

# UNIFORM DISTRIBUTION

```python
Age=np.random.uniform(20,30,1000)
plt.hist(Age)
plt.title("Uniform Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
